```python
"""
import libraries, insert paths and set matplotlib backend
"""

# import python libraries 
import os
import sys
from os import listdir
from os.path import isfile, join

# import data libraries
import numpy as np
import matplotlib.pyplot as plt
import soundfile as sf
from scipy import signal

# set backend
%matplotlib inline

# import library
from music_code import music_code

# initialize
m = music_code.MusicCode(120)


"""
visualization class  
"""

class Wave_Visual:
    
    def __init__(self):
        
        # add integer auto increment to file name
        file_name = 'WavArt_1.png'
        all_names = os.listdir(sys.path.insert(1, r'C:\\Users\\wesle\\Desktop\\repository\\music-code-master\\program files\\images'))
        i=1
        while file_name in all_names:
            i+=1
            file_name_items = file_name[:-4].split('_')
            iteration_num =int(file_name[:-4].split('_')[-1])
            iteration_num+=1
            file_name_items[-1] = str(iteration_num)
            file_name = '_'.join(file_name_items)+'.png'
        
        self.file_name = file_name    
        self.waveform = None
        self.fig = None
    
    # plot waveform
    def plot_wave(self):
        """
        Create wave & plot visual
        """
        
        self.fig = plt.figure(dpi=1200)
        #fig.set_size_inches(5,5)
        ax= self.fig.add_axes([0.1,0.1,0.8,0.8])
        self.fig.patch.set_visible(False)
        ax.axis('off')
        
        i = 0
        curve_values = list(np.linspace(1,250,500))
        for fade_value in list(np.linspace(1/1024,1/32,500)):
            # Wave Designer
            curve_value = curve_values[i]
            self.waveform = m.create_wave(['C1','C6'],'sine-square',1/32, wt_pos=1).fade(fade_in=fade_value, fade_out=fade_value, in_curve=curve_value, out_curve=curve_value)+.5
            ax.plot(self.waveform)
            i+=1
            

    def save_image(self):
        # export hd image             
        png_file_path = "C:\\Users\\wesle\\Desktop\\repository\\music-code-master\\program files\\images\\"+str(self.file_name)
        self.fig.savefig(png_file_path,transparent=True)
        with open(png_file_path, 'w') as outfile:
            self.fig.canvas.print_png(png_file_path)

    def create_visual(self):
        
        self.plot_wave()
        self.save_image()
```

```python
w = Wave_Visual()

w.create_visual()
```
