# Microtonal Harmony
The chord and arpeggio methods are both capable of operating in microtonal pitch systems. In conventional music, the octave is split into 12 equal steps. What if you divided the octave into 15 steps? Or 24 steps? Each tonal system gives us access to a unique set of pitches we can use to compose with. Some of these sounds might feel "out of tune" to you. This is a universal human bias. Almost all of us grew up hearing exclusively 12 tone music. We can go beyond this predisposition and explore entirely new systems of harmony. Let's look at a few examples!

```python
# import library
from music_code import music_code

# initialize
m = music_code.MusicCode(120)
```

```python
# 15 tone major chord... sounds weird right?

micro_chord = m.chord('maj triad', 'C4', 'tri', duration=1, tonal_system=15)

micro_chord.bounce()
```

```python
# 9 tone major six chord

micro_chord = m.chord('maj 6', 'C4', 'tri', duration=1, tonal_system=9)

micro_chord.bounce()
```

```python
# 14 tone minor six arpeggio

micro_arp = m.arpeggio('min 11', 'C3', 'tri', rhythm=1/16,duration=2, tonal_system=11)

micro_arp.bounce()
```

```python
# create phasing by increasing tonal_system several orders of magnitude

phase_wave_1 = m.chord('min 11', 'C3', 'saw1', duration=2, tonal_system=1100).fade(fade_in=1/2)

phase_wave_1.bounce()
```

```python
# more phase examples 

phase_wave_1 = m.chord('min 11', 'E1', 'saw1', duration=2, tonal_system=1100).fade(fade_in=1/2)

phase_wave_1.bounce()
```
