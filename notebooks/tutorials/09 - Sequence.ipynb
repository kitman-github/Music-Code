# Sequence
The sequence method is an efficient way to create wave sequences such as melodies and bass lines. Let's take a look at a few examples.

```python
# import library
from music_code import music_code

# initialize
m = music_code.MusicCode(120)
```

### Sequence

###### required parameters
1. note_labels
    - type: list
    - This is a list of all pitches you want in your waveform. Each item in the list should be either a float representing a frequency (440.0), or string representing a musical note label ('C3') 
2. wave_type
    - type: string
    - waveform shape ('sine', 'tri', 'saw1', 'saw2', 'square')
3. rhythm
    - type: float
    - length of each note in sequence defined by relative rhythmic value (1/4, 1/8, 2 etc.)
    
3. duration
    - type: float
    - total length of sequence defined by relative rhythmic value (1/4, 1/8, 2 etc.)

```python
# input note labels, wave_type, rhythm and duration
sequence_1 = m.sequence(['C3', 'E3', 'G3', 'B3'], 'tri', rhythm=1/8, duration=1)

sequence_1.bounce()
```

```python
# to create rest in sequence
sequence_2 = m.sequence(['C3', 'rest', 'G3', 'B3'], 'tri', rhythm=1/8, duration=1)

sequence_2.bounce()
```

```python
# you can pass wave objects into the sequencer, for example include a rest
sequence_3 = m.sequence(['C3', m.rest(1/4), 'G3', 'B3'], 'tri', rhythm=1/8, duration=1)

sequence_3.bounce()
```

```python
# input wave objects into the sequencer

# create wave
wave1 = m.create_wave(['C2','E3','G3'],'tri',1/4)

# create sequence with included wave object
sequence_4 = m.sequence(['C3', 'F3', wave1, 'G2'], 'tri', rhythm=1/4, duration=2)

# bounce audio
sequence_4.bounce()
```

```python
# input list of frequencies
sequence_4 = m.sequence([440, 550, 660, 770], 'tri', rhythm=1/16, duration=1)

sequence_4.bounce()
```

#### arp_type - this parameter sets the sequence order. 
    - 'up'
    - 'down'
    - 'up-down'
    - 'down-up'

```python
# up-down arpeggio style
sequence_updown = m.sequence(['C3', 'E3', 'G3', 'B3'], 'tri', rhythm=1/8, duration=1, arp_type='up-down')

sequence_updown.bounce()
```

### ADSR 
use fade_in, fade_out, in_curve and out_curve to adjust ADSR parameters

```python
sequence_5 = m.sequence(['C3', 'rest', 'B3', 'G3'], 'tri', rhythm=1/8, duration=1, fade_in=1/8)

sequence_5.bounce()
```

```python
sequence_6 = m.sequence(['C3', 'rest', 'B3', 'G3'], 'tri', rhythm=1/8, duration=1, fade_out=1/8)

sequence_6.bounce()
```

```python
# adjust out_curve value
sequence_7 = m.sequence(['C3', 'rest', 'B3', 'G3'], 'tri', rhythm=1/8, duration=1, fade_out=1/8, out_curve=5).vol(2)

sequence_7.bounce()
```

```python
# set fade_out value to be <= than rhythm value
sequence_8 = m.sequence(['C3', 'rest', 'B3', 'G3'], 'tri', rhythm=1/8, duration=1, fade_out=1/8, out_curve=5)

sequence_8.bounce()
```

# More examples

```python
# use different values for fade_in and fade_out
sequence_9 = m.sequence(['C2','B2','D3','F#3'], 'sine', rhythm=1/16, duration=1, fade_in=1/32, fade_out=1/16, out_curve=7).vol(13)

sequence_9.bounce()
```

### BASS

```python
# bass example
m.sequence(['E0','E0','E0','rest','E0','E0','E0','F0'], 'tri', rhythm=1/8, duration= 2, fade_in=1/8, fade_out=1/16).vol(4).bounce()
```
