# LFO

```python
# import library
from music_code import music_code

# initialize
m = music_code.MusicCode(120)
```

```python
# create wave
waveform = m.create_wave(['C4'], 'sine', 2)

# apply LFO, .1 Hz sine wave
waveform = waveform.LFO(.1, 'sine')

# WAV audio
waveform.bounce()
```

```python
# create chord
waveform = m.chord('maj 9', 'F3', 'saw1', 2)

# apply .2 Hz tri wave LFO
waveform = waveform.LFO(.2, 'tri')

# apply 1 bar fade in
waveform = waveform.fade(fade_in=1,fade_out=1/8).vol(3)

# WAV audio
waveform.bounce()
```

```python
### Apply multiple LFOs
```

```python
# create arpeggio
waveform = m.arpeggio('maj 9', 'F3', 'tri', rhythm=1/16, duration=1,fade_in=1/16).vol(2)

# LFO 1
waveform = waveform.LFO(.1, 'sine').LFO(.2, 'sine').loop(2).vol(4)

waveform.bounce()
```
