# Join/Add Waves
Using time_edit, we can adjust the duration of any waveforms. We can also join multiple waveforms together vertically and horizontally. Let's look at some examples. First, run your set up code below to initialize MusiCode.

```python
# import library
from music_code import music_code

# initialize
m = music_code.MusicCode(120)
```

### Time Edit
We can edit the total duration of a waveform. There are two time editing modes, <b>relative</b> and <b>absolute</b>.

##### Relative (default)
Relative mode works with musical rhythms. We can edit waveform duration to be a quarter note, eight note, 2 measures or any other sensible rhythmic value. Remember that relative mode will change depending on which BPM you have set globally.

##### Absolute
Absolute mode works with seconds. We can edit waveform duraton to be 5 seconds using absolute mode. Let's see examples of both.

```python
# create unedited waveform
waveform = m.create_wave(['C3','E3','G3'],'tri',1)

# view waveform
waveform.view('waveform')
```

```python
# use time_edit to lengthen waveform duration by adding silence

# make waveform 2 bars in duration
waveform.time_edit(2)

# view waveform
waveform.view('waveform')
```

```python
# use time_edit to shorten waveform duration by removing part of the waveform

# remove silence
waveform.time_edit(1)

# graph
waveform.view('waveform')
```

### Time

```python
# print waveform total duration in seconds

waveform.time()
```

### Join Waves
Join together seperate waveforms into a sequence (horizontal)

##### Melody

```python
# create individual notes
note_1 = m.create_wave(['C3'], 'tri', 1/8)
note_2 = m.create_wave(['F3'], 'tri', 1/8)
note_3 = m.create_wave(['B3'], 'tri', 1/8)
note_4 = m.create_wave(['G3'], 'tri', 1/8)

# join notes into melody
melody = m.join_waves((note_1, note_2, note_3, note_4))

# check melody waveform duration
melody.time()

# bounce audio
melody.bounce()
```

##### Harmony

```python
# create chords
c_major = m.create_wave(['C3','E3','G3'],'tri',1)
f_major = m.create_wave(['C3','F3','A3'],'tri',1)
a_minor = m.create_wave(['C3','E3','A3'],'tri',1)
g_major = m.create_wave(['B2','D3','G3'],'tri',1)

# join chords into progression
chord_progression = m.join_waves((c_major, f_major, a_minor, g_major))

# check waveform duration
chord_progression.time()

# bounce audio
chord_progression.bounce()
```

### Add Waves

```python
# create two individual notes
note_C3 = m.create_wave(['C3'],'sine',1)
note_G3 = m.create_wave(['G3'],'sine',1)

# add them together to create a perfect fifth harmonic interval
interval = m.add_waves((note_C3, note_G3))

# check waveform duration
interval.time()

interval.bounce()
```

### Value Error: Operands don't match
All waveforms used in the add_waves statement must be the same duration or it will cause a value error. Here is an example of what causes the error and how to fix it.

```python
# 1 bar duration
note_C3 = m.create_wave(['C3'],'sine',1)

# 2 bar duration
note_G3 = m.create_wave(['G3'],'sine',2)

# try to add different sized waves together,
interval = m.add_waves((note_C3, note_G3))
```

### Solution

```python
# run time() for each variable in your add_waves statement. all numbers should be the same, or error will occur.
note_C3.time()
note_G3.time()
```

```python
# Make both waveforms same duration, 2 bars

# 2 bar duration
note_C3 = m.create_wave(['C3'],'sine',2)

# 2 bar duration
note_G3 = m.create_wave(['G3'],'sine',2)

# run time() for each variable. all numbers below should be the same, or error will occur
note_C3.time()
note_G3.time()
```

```python
# now it works
interval = m.add_waves((note_C3, note_G3))

interval.bounce()
```
