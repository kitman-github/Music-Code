# Looping
This allows you to repeat musical phrases and create recurring patterns.

```python
# import library
from music_code import music_code

# initialize
m = music_code.MusicCode(120)
```

```python
# create two seperate chords
c_major = m.create_wave(['C3','E3','G3'],'tri',1/4)
f_major = m.create_wave(['C3','F3','A3'],'tri',1/4)

# join chords into one waveform
chords = m.join_waves((c_major,f_major))

# bounce audio
chords.bounce()
```

```python
# loop chords 4 times
chords.loop(4)

# bounce audio
chords.bounce()
```

```python
# create C4 wave (1/100 duration)
vibration = m.create_wave(['C3'],'sine',1/100)

# loop 100 times
vibration.loop(100)

# WAV audio
vibration.bounce()
```
