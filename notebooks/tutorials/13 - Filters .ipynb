# Filters

```python
# import library
from music_code import music_code

# initialize
m = music_code.MusicCode(120)
```

```python
# create waveform with no filter
no_filter = m.create_wave(['C4','E4','G4','B4'],'saw1',2)

no_filter.bounce()
```

```python
# filter out all frequencies over 1000Hz
low_pass = m.create_wave(['C4','E4','G4','B4'],'saw1',2).LPF(1000)

low_pass.bounce()
```

```python
# filter out all frequencies under 2000Hz

# if you receive a red FutureWarning when using high_pass, it is not an error and should still work.

high_pass = m.create_wave(['C4','E4','G4','B4'],'saw1',2).HPF(2000)

high_pass.bounce()
```

```python
# bright timbre
high_pass.view('spectro', freq_range=(100,10000))
```

```python
# dark timbre
low_pass.view('spectro', freq_range=(100,10000))
```
