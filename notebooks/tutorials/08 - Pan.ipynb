# Pan
This allows you to utilize the full stereo field.

```python
# import library
from music_code import music_code

# initialize
m = music_code.MusicCode(120)
```

```python
# center mono
wave_mono = m.create_wave(['C3'], 'tri', 1/2).pan('C')

wave_mono.bounce()
```

```python
# pan right
wave_right = m.create_wave(['C3'], 'tri', 1/2).pan('R50')

wave_right.bounce()
```

```python
# pan left
wave_left = m.create_wave(['C3'], 'tri', 1/2).pan('L50')

wave_left.bounce()
```

```python
# pan right (1-50 range)
wave_right = m.create_wave(['C3'], 'saw1', 1/4).pan('R25')

wave_right.bounce()
```

```python
# pan left (1-50 range)
wave_left = m.create_wave(['C3'], 'saw1', 1/4).pan('L33')

wave_left.bounce()
```

```python
# pan left (1-50 range)
wave_left = m.create_wave(['C3'], 'saw1', 1/4).pan('L5')

wave_left.bounce()
```
