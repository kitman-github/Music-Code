# Create Wave & Rest

```python
# import library
from music_code import music_code

# initialize
m = music_code.MusicCode(120)
```

### Create Wave
The create_wave method is a core building block of MusiCode. Notice that the create_wave statement below starts with m.
###### required parameters
1. note_labels
    - type: list
    - This is a list of all pitches you want in your waveform. Each item in the list should be either a float representing a frequency (440.0), or string representing a musical note label ('C3') 
2. wave_type
    - type: string
    - waveform shape ('sine', 'tri', 'saw1', 'saw2', 'square')
3. duration
    - type: float
    - total length of waveform defined by relative rhythmic value (1/4, 1/8, 2 etc.)

```python
# create 440hz sine wave and store in variable named waveform
waveform = m.create_wave(note_labels=[440], wave_type='sine', duration=1/8)

# WAV audio 
waveform.bounce('wave test.wav')
```

```python
# create 50hz sine wave
sub_bass = m.create_wave(note_labels=[50], wave_type='sine', duration=1/8)

# WAV audio (this is a low frequency sound you will not hear on small speakers)
sub_bass.bounce('wave test.wav')
```

```python
# create middle C triangle wave
middle_c = m.create_wave(note_labels=['C3'], wave_type='tri', duration=1/8)

# WAV audio
middle_c.bounce('middle_c.wav')
```

### Rest

```python
# rhythm value input to create silence
m.rest(1/2).bounce()
```

### BPM & Relative Duration
It is important to point out that duration and rhythm in MusiCode V12 are RELATIVE not absolute. This means that instead of duration being defined in seconds, duration is defined in bars. For example, if you set waveform duration to 1/4, that is a quarter note duration. The number of seconds in that quarter note changes depending on the BPM you set. A quarter note at 60 BPM is 1 second and a quarter note at 120BPM is .5 seconds. If you set duration to 4, this means four bars using 4/4 time signature.

```python
# change BPM
m.set_bpm(140)
```

```python
# create c major chord
c_major = m.create_wave(['C3','E3','G3'], 'tri', 1/2)

# WAV audio
c_major.bounce('C major chord.wav')
```

```python
# create F# minor chord
f_sharp_major = m.create_wave(['F#3','A3','C#4'], 'tri', 1/4)

# WAV audio
f_sharp_major.bounce('F# minor chord.wav')
```

```python
# create Ab major chord
a_flat_major = m.create_wave(['Ab3','C3','Eb3'], 'tri', 1/2)

# WAV audio
a_flat_major.bounce('Ab major chord.wav')
```

### Volume



```python
# 2x volume increase
waveform = m.create_wave(['C3'], 'tri', 1/2).vol(2)

# 1/2 volume decrease
waveform = m.create_wave(['C3'], 'tri', 1/2).vol(1/2)
```

### Wavetable Position
Wt_pos adjusts the timbre of a waveform. Lets look at the difference between the following waveforms:

```python
# set wt_pos = 7
waveform = m.create_wave(['C0'], 'sine', 1/8, wt_pos=7)

waveform.view('waveform')
```

```python
# the following two create_wave statements are equivalent

# all amplitude values in wave array are raised to the power of 7
waveform = m.create_wave(['C0'], 'sine', 1/8) ** 7

# wt_pos=7 is the same as waveform ** 7
waveform = m.create_wave(['C0'], 'sine', 1/8, wt_pos=7)
```

```python
# set wt_pos = 77
waveform = m.create_wave(['C0'], 'sine', 1/8, wt_pos=77)

waveform.view('waveform')
```

```python
# set wt_pos = 77777
waveform = m.create_wave(['C0'], 'sine', 1/8, wt_pos=77777)

waveform.view('waveform')
```

```python
# set wt_pos = 777777
waveform = m.create_wave(['C0'], 'sine', 1/8, wt_pos=777777)

waveform.view('waveform')
```

```python
# the next three waves are sine waves of the same frequency... what is the difference in timbre of these waveforms?

waveform1 = m.create_wave(['C3'], 'sine', 1/2)
waveform2 = m.create_wave(['C3'], 'sine', 1/2, wt_pos=99).vol(.4)
waveform3 = m.create_wave(['C3'], 'sine', 1/2, wt_pos=999).vol(.4)
```

```python
waveform1.bounce('waveform1.wav')
```

```python
waveform2.bounce('waveform2.wav')
```

```python
waveform3.bounce('waveform3.wav')
```
