## ADSR
Attack, decay, sustain and release are how we shape the volume of a sound. We can shape the amplitude of the waveforms we can create by utilizing fades. Here is an example.

- Fades have four parameters:
    - fade_in
    - fade_out
    - in_curve
    - out_curve

```python
# linear fade in
m.create_wave(['C3'],'sine',1).fade(fade_in=1/2).bounce()
```

```python
# linear fade out
m.create_wave(['C3'],'sine',1).fade(fade_out=1/2).bounce()
```

```python
# linear fade in and out
m.create_wave(['C3'],'sine',1).fade(fade_in = 1/2, fade_out=1/2).bounce()
```

```python
# exp fade in and out
m.create_wave(['C3'],'sine',1).fade(fade_in = 1/2, in_curve=6, fade_out=1/2, out_curve=6,).bounce()
```

```python
# exponential fade in 
m.create_wave(['C3'],'sine',1).fade(fade_in = 1/2, in_curve=5).bounce()
```

```python
# exponential fade out
m.create_wave(['C3'],'sine',1).fade(fade_out = 1/2, out_curve=5).bounce()
```

```python
# combine different fade ins and outs
m.create_wave(['C3'],'sine',1).fade(fade_in = 1/2, fade_out = 1/2, in_curve=9).bounce()
```

```python
m.create_wave(['C3'],'sine',1).fade(fade_in=1/4, fade_out=3/4, in_curve=3, out_curve=4).bounce()
```

```python
m.create_wave(['C3'],'sine',1).fade(fade_in=3/4, fade_out=1/4, in_curve=3).bounce()
```
