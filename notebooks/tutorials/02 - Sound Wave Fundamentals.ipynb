# Sound Wave Fundamentals

Waves are a fundamental pattern of nature and can exist in many forms including sound, light and the elctromagnetic spectrum. 

Sounds are waves. We can use code to genreate waveforms. The basic waveform shapes are sine, triangle, sawtooth and square. These <b>waveforms</b> can be generated using equations which we can access through the scipy library. Each waveform produces a slightly different <b>timbre</b>. For more info on the physics of sound waves check out: [wave physics](https://www.d.umn.edu/~mharvey/th1551terminology.html)

<b>Fun Fact:</b> For the first time ever, scientists recently detected <b>gravitational waves</b>, which are literal ripples in the fabric of spacetime. Spacetime stretches, warps and distorts as matter moves through it. This causes waves to propogate. If the wave amplitude is large enough it can be detected using a laser interferometer. Gravitational waves were predicted by Einsteins theory of relativity and only recently were proven to exist. Wow! For more info on gravitational waves go to: [gravitational waves](http://www.physics.org/article-questions.asp?id=138)

#### Key Sound Wave Terms
1. <b>waveform</b> - pattern of pressure variation (amplitude) in the time domain
2. <b>amplitude</b> - the loudness of a waveform, measured in decibels (Db)
3. <b>frequency</b> - number of peaks (cycles) in a given period of time (also known as pitch or "musical notes") Measured in Hertz (Hz).
4. <b>timbre</b> - the character or quality of a sound.

```python
# import libraries for graphing 
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
%matplotlib inline

# add scripts folder to path
import sys
sys.path.insert(1, r'C:\Users\wesle\Desktop\repository\music-code-master\scripts')

# initialize musicode
from music_code import MusicCode
m = MusicCode(120)
```

```python
# helper function for graphing waveforms
def graph(waveform,title):
    plt.style.use('seaborn-dark-palette')
    fig, axs = plt.subplots(1,figsize=(700/96, 200/96), dpi=96)
    fig.suptitle(title)
    axs.plot(waveform, color='mediumblue')
    plt.show()
```

```python
# define total number of samples... there are 44100 samples in 1 second of 44.1KHz audio
samples = np.arange(1000)

# audio sample rate (Hz)
sample_rate = 44100

# frequency/pitch
freq = 440 # A4 
```

```python
# generate sine wave using equation
sine_wave = np.sin(1 * np.pi * freq * samples / sample_rate)

# graph
graph(sine_wave,'sine wave')
```

```python
# generate sine wave
sine_wave = m.create_wave(['C4'],'sine', 1/2)

# play WAV audio
sine_wave.bounce()
```

```python
#  generate triangle wave using equation
tri_wave = signal.sawtooth(t=1 * np.pi * freq * samples / sample_rate,width=.5)*.5

# graph
graph(tri_wave,'triangle wave')
```

```python
# generate triangle wave using create_wave
tri_wave = m.create_wave(['C4'],'tri', 1/2)

# play WAV audio
tri_wave.bounce()
```

```python
# generate sawtooth 1 wave using equation
saw1_wave = signal.sawtooth(t=1 * np.pi * freq * samples / sample_rate,width=1)

# graph
graph(saw1_wave,'sawtooth 1 wave')
```

```python
# generate sawtooth 1 wave using create_wave
saw1_wave = m.create_wave(['C4'],'saw1', 1/2)

# play WAV audio
saw1_wave.bounce()
```

```python
# generate sawtooth 2 wave using equation
saw2_wave = signal.sawtooth(t=1 * np.pi * freq * samples / sample_rate,width=0)

# graph
graph(saw2_wave,'sawtooth 2 wave')
```

```python
# generate sawtooth 2 wave using create_wave
saw2_wave = m.create_wave(['C4'],'saw2', 1/2)

# play WAV audio
saw2_wave.bounce()
```

```python
# generate square wave using equation
square_wave = signal.square(t=1 * np.pi * freq * samples / sample_rate)

# graph 
graph(square_wave,'square wave')
```

```python
# generate square wave using create_wave
square_wave = m.create_wave(['C4'],'square', 1/2)

# play WAV audio
square_wave.bounce()
```
