# Arpeggio

```python
# import library
from music_code import music_code

# initialize
m = music_code.MusicCode(120)
```

```python
m.set_bpm(120)

# major thirteen arpeggio
arpeggio_1 = m.arpeggio('maj 6 9', 'F3', 'sine-tri', rhythm=1/16,duration=2, fade_out=1/24).vol(1.3)

arpeggio_1.bounce()
```

```python
# minor seven arpeggio with fade_in
arpeggio_2 = m.arpeggio('min 7', 'G2', 'saw-square', rhythm=1/16,duration=1, fade_in=1/16)

arpeggio_2.bounce()
```

```python
# minor seven arpeggio with fade_in
arpeggio_3 = m.arpeggio('northern lights chord', 'C3', 'sine-tri', rhythm=1/16,duration=1, fade_out=1/16, out_curve=3).vol(3)

arpeggio_3.bounce()
```
